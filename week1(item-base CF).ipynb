```python
ratings.drop('timestamp', axis=True, inplace=True)
ratings.head(10)
```

```python
r_m = ratings.merge(movies, on='movieId')
r_m.head(5)
```

```python
matrix= pd.pivot_table(r_m, index=['title'],columns=['userId'],values='rating').fillna(0)
matrix.head(5)
```

```python
matrix_= pd.pivot_table(r_m, index=['userId'],columns=['title'],values='rating').fillna(0)
matrix_.head(5)
```

우선 블로그(https://lsjsj92.tistory.com/568)를 따라한 item based CF는 `matrix` 사용할 예정
- index = movie title
- columns = user id

```python
# 코싸인 similarity 라이브러리 불러오기 (skikit learn에서 제공)
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_cosine = cosine_similarity(matrix)
item_cosine
```

```python
item_cosine.shape
```

```python
item_cosine[0,0:5]
```

```python
item_cosine = pd.DataFrame(data=item_cosine, index=matrix.index, columns = matrix.index)
item_cosine
```

```python
item_cosine.columns
```

```python
item_cosine.index
```

```python
def recommendation1(title):
    return item_cosine[title].sort_values(ascending=False)[1:6] #max(cosine) 값은 자기자신과의 상관관계니까 지움
```

```python
recommendation1('"Great Performances" Cats (1998)') 
```
